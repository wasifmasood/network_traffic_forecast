### Encode the cell names

```python
df_raw.cell_key = LabelEncoder().fit_transform(list(df_raw.cell_key))        
```

```python
df_raw['dt'] = df_raw['dt'].astype(str)
df_raw['quarter_hour'] = df_raw['quarter_hour']
df_raw['DateTime'] = df_raw['dt'] +' ' + df_raw['quarter_hour']
df_raw['DateTime'] = pd.to_datetime(df_raw['DateTime'], format='%Y%m%d %H:%M:%S.%f')
df_raw = df_raw.sort_values(['cell_key','DateTime'], ascending=True)
```
